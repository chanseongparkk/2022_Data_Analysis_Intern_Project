```python
# 동 feature 만들기
def 동(df, feature):
 동 = []
 for i in range(len(df)):
   동_list = df[feature].str.split()[i][2]
   동.append(동_list)
 df["동"] = 동

# 구 feature 만들기
def 구(df, feature):
 구 = []
 for i in range(len(df)):
   구_list = df[feature].str.split()[i][1]
   구.append(구_list)
 df["구"] = 구
 
# 월 feature 만들기
def 월(df, feature):
 월 = []
 for i in range(len(df)):
   월_list = df[feature].str.split()[i][1]
   월.append(월_list)
 df["월"] = 월

# 년 feature 만들기
def 년(df, feature):
 년 = []
 for i in range(len(df)):
   년_list = df[feature].str.split()[i][0]
   년.append(년_list)
 df["년"] = 년
```

#### 1. 바닥신호등 설치 전 후 사고 건수 비교

```python
gn18_19 = pd.read_csv("/content/drive/MyDrive/4조 프로젝트 정리/사고데이터 및 바닥신호등 폴더/강남,광진,동대문구/강남구_18~19.csv", encoding = "euc-kr")
gn20_21 = pd.read_csv("/content/drive/MyDrive/4조 프로젝트 정리/사고데이터 및 바닥신호등 폴더/강남,광진,동대문구/강남구_20~21.csv", encoding = "euc-kr")
gn_tr = pd.read_csv("/content/drive/MyDrive/4조 프로젝트 정리/사고데이터 및 바닥신호등 폴더/강남,광진,동대문구/강남구바닥신호등.csv", encoding = "euc-kr")
display(gn18_19.head(3), gn20_21.head(3), gn_tr.head(3))
```

```python
동(gn18_19, "시군구")
동(gn20_21, "시군구")
동(gn_tr, "소재지지번주소")
```

```python
# 도로명주소로 나와있는 동으로 처리
gn_tr["동"][17] = "역삼동"
gn_tr["동"][18] = "도곡동"
gn_tr["동"][18] = "역삼동"
```

```python
# 강남구 바닥신호등 설치 동 리스트
gn_tr_list = list(gn_tr.동.unique())
gn_tr_list
```

```python
gn18_19["설치여부"] = np.nan

for i in range(len(gn18_19)):
  if gn18_19["동"][i] in gn_tr_list:
    gn18_19["설치여부"][i] = 1
  else:
    gn18_19["설치여부"][i] = 0

print(gn18_19["설치여부"].value_counts())
```

```python
# 설치되어있는 동의 18-19년 사고 횟수

gn18_19[gn18_19["설치여부"] == 1].shape[0]
```

```python
gn20_21["설치여부"] = np.nan

for i in range(len(gn20_21)):
  if gn20_21["동"][i] in gn_tr_list:
    gn20_21["설치여부"][i] = 1
  else:
    gn20_21["설치여부"][i] = 0

print(gn20_21["설치여부"].value_counts())
```

```python
# 설치되어있는 동의 20-21년 사고 횟수

gn20_21[gn20_21["설치여부"] == 1].shape[0]
```

```python
# 전 후 차이 
print(gn18_19[gn18_19["설치여부"] == 1].shape[0] - gn20_21[gn20_21["설치여부"] == 1].shape[0])
```

#### 2. t-test

```python
월(gn18_19, "사고일시")
년(gn18_19, "사고일시")
월(gn20_21, "사고일시")
년(gn20_21, "사고일시")
```

```python
display(gn18_19.iloc[:,-2:],gn20_21.iloc[:,-2:])
```

```python
list_gn_1819 = np.array(gn18_19.groupby(["년","월"]).count().iloc[:,0])
list_gn_2021 = np.array(gn20_21.groupby(["년","월"]).count().iloc[:,0])
```

```python
# 등분산성 검정
stats.levene(list_gn_1819, list_gn_2021)
# p_value 0.216으로 유의수준 5% 이내에서 등분산성이라는 귀무가설을 채택한다.
```

```python
# 정규성 검정
print(stats.shapiro(list_gn_1819),"\n",stats.shapiro(list_gn_2021))
# 18_19 데이터는 p_value 0.604, 20_21 데이터는 p_value 0.217으로 유의수준 5% 이내에서 정규성을 만족한다는 귀무가설을 채택한다.
```

```python
# 단측검정
stats.ttest_ind(list_gn_1819,list_gn_2021,alternative='greater')

# H0: 바닥신호등 설치 전과 후의 사고횟수 차이가 없다.
# H1: 바닥신호등 설치 전의 사고횟수가 설치 후의 사고횟수보다 크다.

# p_value 0.007***로 유의수준 5% 이내에서 설치 전과 후가 같다는 귀무가설을 기각하고 설치 전이 후보다 사고 횟수가 많다는 대립가설을 채택한다. 
```

#### 3. 바닥신호등 설치 동 전 후 사고 감소량 계산 

```python
# 바닥신호등 설치된 동만 추출

gn18_19_ins = gn18_19[gn18_19["설치여부"] == 1]
gn20_21_ins = gn20_21[gn20_21["설치여부"] == 1]
```

```python
# 동 별 바닥신호등 설치 전후 사고건수 추출

bef_acc_gn = gn18_19_ins.value_counts("동") # 설치 전 사고데이터
aft_acc_gn = gn20_21_ins.value_counts("동") # 설치 후 사고데이터
```

```python
# 동 별 바닥신호등 설치 전후 사고건수 결합 
gn_acc = pd.concat([pd.DataFrame(bef_acc_gn,columns = ["before"]),pd.DataFrame(aft_acc_gn,columns = ["after"])],axis = 1)
gn_acc
```

```python
# 감소율 변수 생성
gn_acc["decrease_rate"] = 1 - (gn_acc["after"] / gn_acc["before"] )
gn_acc
```

```python
c = pd.concat([pd.DataFrame(gn18_19[gn18_19["설치여부"] == 0].value_counts("동"),columns = ["before"]),pd.DataFrame(gn20_21[gn20_21["설치여부"] == 0].value_counts("동"),columns = ["after"])],axis = 1)
1 - (c["after"] / c["before"] )
```

```python
# 그룹 별 경계값(mu: before 중앙값보다 큰 데이터, ml: before 중앙값보다 작은 데이터)
mu_gr_a = decrease_rate[decrease_rate.before >= decrease_rate.before.median()].decrease_rate.quantile(0.8)
mu_gr_b = decrease_rate[decrease_rate.before >= decrease_rate.before.median()].decrease_rate.quantile(0.3) #중앙값 넘는애들30-80 -> B
ml_gr_b = decrease_rate[decrease_rate.before < decrease_rate.before.median()].decrease_rate.quantile(0.8)  #안넘는애들 30-100 -> B
```

```python
decrease_rate["group"] = np.nan

for i in range(decrease_rate.shape[0]):
 if (decrease_rate["decrease_rate"][i] >= mu_gr_a) & (decrease_rate["before"][i] >= decrease_rate["before"].median()) :
   decrease_rate["group"][i] = "A"
 elif (decrease_rate["decrease_rate"][i] >= mu_gr_b) & (decrease_rate["before"][i] >= decrease_rate["before"].median()) :
   decrease_rate["group"][i] = "B"
 elif (decrease_rate["decrease_rate"][i] >= ml_gr_b) & (decrease_rate["before"][i] < decrease_rate["before"].median()):
   decrease_rate["group"][i] = "B"
 else:
   decrease_rate["group"][i] = "C"
```
