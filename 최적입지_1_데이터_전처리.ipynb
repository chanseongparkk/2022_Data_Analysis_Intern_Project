```python
import pandas as pd
import numpy as np
import math
```

```python
# 동 feature 만들기
def 동(df, feature):
 동 = []
 for i in range(len(df)):
   동_list = df[feature].str.split()[i][2]
   동.append(동_list)
 df["동"] = 동

# 구 feature 만들기
def 구(df, feature):
 구 = []
 for i in range(len(df)):
   구_list = df[feature].str.split()[i][1]
   구.append(구_list)
 df["구"] = 구

 # 시 feature 만들기
def 시(df, feature):
 시 = []
 for i in range(len(df)):
   시_list = df[feature].str.split()[i][0]
   시.append(시_list)
 df["시"] = 시

 # 월 feature 만들기
def 월(df, feature):
 월 = []
 for i in range(len(df)):
   월_list = df[feature].str.split()[i][1]
   월.append(월_list)
 df["월"] = 월

# 년 feature 만들기
def 년(df, feature):
 년 = []
 for i in range(len(df)):
   년_list = df[feature].str.split()[i][0]
   년.append(년_list)
 df["년"] = 년
```

### 서울시 차대사람 사고데이터 전처리

```python
acc = pd.read_csv("/content/drive/MyDrive/4조 프로젝트 정리/2019원본데이터/2019년도 교통사고 데이터.csv")
acc
```

```python
# 동 변수 생성

동(acc, "시군구")
```

```python
# 시간 변수 생성

acc["시간"] = np.nan

for i in range(acc.shape[0]):
  acc["시간"][i] = acc["사고일시"].str.split()[i][3]

acc
```

```python
# 주야구분 변수 생성

acc["주야구분"] = np.nan

for i in range(acc.shape[0]):
 if acc["시간"][i] in ["20시","21시","22시","23시","00시","01시","02시","03시","04시","05시","06시"]:
   acc["주야구분"][i] = "야간"
 else :
   acc["주야구분"][i] = "주간"

acc
```

```python
acc_count = acc.groupby(["동","주야구분"]).size()
acc_count = acc_count.reset_index()
acc_count = acc_count.rename(columns = {0:"사고횟수"})
acc_count
```

```python
acc_day = acc_count[acc_count.주야구분 == "주간"].rename(columns = {"사고횟수":"주간사고횟수"})
acc_night = acc_count[acc_count.주야구분 == "야간"].rename(columns = {"사고횟수":"야간사고횟수"})
acc_count_final = pd.merge(acc_day,acc_night,how = "outer" ,on = "동").iloc[:,[0,2,4]]
acc_count_final
```

```python
# 사고횟수가 na로 들어가 있다는 것은 사고 횟수가 없음을 뜻한다. 따라서 0으로 채워준다.

acc_count_final[acc_count_final.주간사고횟수.isna() == True] = acc_count_final[acc_count_final.주간사고횟수.isna() == True].replace(np.nan,0)
acc_count_final[acc_count_final.야간사고횟수.isna() == True] = acc_count_final[acc_count_final.야간사고횟수.isna() == True].replace(np.nan,0)
acc_count_final.isna().sum() # na값이 0으로 잘 대체되었다.
```

```python
acc_count_final["야간사고비율"] = acc_count_final["야간사고횟수"] / (acc_count_final["주간사고횟수"] + acc_count_final["야간사고횟수"])
acc_count_final
```

```python
acc_count_final["야간사고비율"].describe()
```

```python
a = acc_count_final.copy()
a = a[a.주간사고횟수>=10]
a.야간사고비율.describe()
```

**야간사고 위험도 기준**

* 주간사고 10건 이상 & 야간사고비율 37% 이상 -> A그룹
* 주간사고 10건 이상 & 야간사고비율 25% 이상 -> B그룹
* 주간사고 10건 이하 or 주간사고 10건 이상 & 야간사고비율 25% 이하 -> C그룹


```python
# 야간사고 위험도 변수 생성

acc_count_final["야간사고위험도"] = np.nan

for i in range(acc_count_final.shape[0]):
  if (acc_count_final["주간사고횟수"][i] >= 10) & (acc_count_final["야간사고비율"][i] >= 0.37):
    acc_count_final["야간사고위험도"][i] = "A"
  elif (acc_count_final["주간사고횟수"][i] >= 10) & (acc_count_final["야간사고비율"][i] >= 0.25):
    acc_count_final["야간사고위험도"][i] = "B"
  else:
    acc_count_final["야간사고위험도"][i] = "C"

acc_count_final
```

**야간사고 위험도 기준**

* 주간사고 10건 이상 & 야간사고비율 33% 이상 -> A그룹
* 주간사고 10건 이상 & 야간사고비율 20% 이상 -> B그룹
* 주간사고 10건 이하 or 주간사고 10건 이상 & 야간사고비율 20% 이하 -> C그룹


```python
# 야간사고 위험도 변수 생성

acc_count_final["야간사고위험도"] = np.nan

for i in range(acc_count_final.shape[0]):
  if (acc_count_final["주간사고횟수"][i] >= 10) & (acc_count_final["야간사고비율"][i] >= 0.33):
    acc_count_final["야간사고위험도"][i] = "A"
  elif (acc_count_final["주간사고횟수"][i] >= 10) & (acc_count_final["야간사고비율"][i] >= 0.2):
    acc_count_final["야간사고위험도"][i] = "B"
  else:
    acc_count_final["야간사고위험도"][i] = "C"

acc_count_final
```
