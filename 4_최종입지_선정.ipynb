```python
import pandas as pd
import numpy as np
```

### 모델 예측값 분석

```python
# 데이터 불러오기
rf1 = pd.read_csv("모델링_분류라벨/RandomForest_label2.csv")
xgb1 = pd.read_csv("모델링_분류라벨/XGBClassifier_label2.csv")
gnb1 = pd.read_csv("모델링_분류라벨/gnb_label2.csv")
lr1 = pd.read_csv("모델링_분류라벨/LogisticRegression_label2.csv")
svc1 = pd.read_csv("모델링_분류라벨/SVC_label2.csv")
```

```python
# 데이터 결합
df1 = pd.merge(rf1,xgb1,how="outer",on="법정동")
df1 = pd.merge(df1,gnb1,how="outer",on="법정동")
df1 = pd.merge(df1,lr1,how="outer",on="법정동")
df1 = pd.merge(df1,svc1,how="outer",on="법정동")

# 컬럼명 재정의
df1.columns = ["법정동","l_rf","p_rf","l_xgb","p_xgb","l_gnb","p_gnb","l_lr","p_lr","l_svc","p_svc"]
df1 = df1.iloc[:,[0,1,3,5,7,9,2,4,6,8,10]]
all_include1 = df1.dropna() # 5개의 모델 모두에서 A가 나온 지역 필터
all_include1["mean_A_prob"] = all_include1.iloc[:,6:].mean(axis=1) # 5개 모델의 A확률 평균을 변수로 생성
all_include1
```

```python
#df1.to_csv("d.csv", index = False)
```

```python
all_include1.sort_values("mean_A_prob",ascending = False)
```

```python
df2 = df1.copy()
```

```python
df2[["l_rf","l_xgb","l_gnb","l_lr","l_svc"]] = df2.iloc[:,[1,2,3,4,5]].replace(np.nan, "not A")
df2
```

```python
df2["count"] = 0

for i in [1,2,3,4,5]:
  for j in range(df2.shape[0]):
    if df2.iloc[j,i] == "A":
      df2["count"][j] += 1


df2
```

```python
df2 = df2[["법정동","count"]]
df2 = df2.sort_values("count", ascending = False)
df2
```

```python
b1 = pd.read_csv("2021전처리완료데이터/최종데이터.csv")
b2 = pd.read_csv("2021전처리완료데이터/설치안된동.csv")
b2 = b2.rename(columns = {"0":"동"})
b3= pd.merge(b2, b1, how = "left", left_on = "동", right_on = "법정동")
b3 = b3.drop(["동"],axis = 1) # 필요없는 변수 삭제
b3
```

```python
df3 = pd.merge(df2,b3.iloc[:,0],how="outer",on="법정동").replace(np.nan,0)
df3 = df3.sort_values("count",ascending = False)
df3
```
